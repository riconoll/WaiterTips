## A Quick Overview
Let's create a new column that shows the tip percentage of each bill

```python
def percentage (tip, bill): 
     return (tip/bill) * 100
(df['tip_percentage']) = percentage(df['tip'], df['total_bill'])

# Let's divide the tips with the total_bill and multiply by 100.
# This will give us percentage the guest tip
```

```python
pd.set_option('display.max_rows',1000)
# With this code, we can have look of the whole table.
# At quick glance, there's good and bad tippers.
# Let's explore the server's tip by the days
df
```

## Tips by Days
Let's calculate the average tips by percentage broken down to the days the server worked

```python
daySales = df.groupby('day').sum()
# Created a new dataframe
# Group by day and get the sum of the bill, tips and how many guest were served
```

```python
daySales['tip_percentage'] = percentage(daySales['tip'],daySales['total_bill'])
# Since we have a new dataframe will need to recalculate and add the column of the tip_percentage
```

```python
daySales
# With this date presented - it seems the server had a better tip percentage Thursday and Friday
# Though looking at the dataframe, it seems the less guest served the higher the tip percentage is
# This can factor whether there has been a change of the way service went. 
# Example: If the restaurant filled out quickly and service slowed down due to all the guest arrving all at once
# Tips are dependable on the amount of guest served and service.
```

```python
# What about the average tip

for day in df['day'].unique():
    mean = df[df['day'] == day]['tip_percentage'].mean()
    print(f'Day: {day} the mean was: {round(mean,2)}%')
# Friday and Sunday seems to have a higher average of tip per bill.
# Because the data doesn't have the same amount of trasaction per day, the of mean is the suited for the data
```

## Conclusion
Depending on the business whether it is slow or busy, factors that affect tip percentage depend on quality of service. A consistent service may affect the tip percentage to be consistent as well.
Granted the server working a busy day or times will grant more money though the tip percentage decrease. It may be because of service or just how the guest are.

## Male or Female Guest
Can we factor in whether the gender affect the tips?
Now keep in mind, our server is Male

```python
df.groupby('sex').size().plot(kind='bar') 
# Majority of the guest were male
```

```python
genSales = df.groupby('sex').mean()
genSales['tip_percentage'] = percentage(genSales['tip'],genSales['total_bill'])
```
